```python
test_data= """...#......
.......#..
#.........
..........
......#...
.#........
.........#
..........
.......#..
#...#....."""
```

```python
N=1+test_data.count('\n')
test_list=[int(c) for line in test_data.replace('.','0').replace('#','1').split('\n')  for c in line]
np_test=np.array(np.array(test_list).reshape(N,N))
```

```python
np_test.shape
```

```python
np_test
```

```python
row_sum=np.sum(np_test,axis=1)
col_sum=np.sum(np_test,axis=0)

zero_rows=np.where(row_sum==0)[0]
zero_cols=np.where(col_sum==0)[0]
```

```python
zero_rows
```

```python
cols=np.hsplit(np_test,N)
for c in reversed(zero_cols):
    cols.insert(c,np.zeros(N).reshape(-1,1))
cols_added=np.hstack(cols)

rows=np.vsplit(cols_added,N)
for r in reversed(zero_rows):
    rows.insert(r,np.zeros(rows[0].shape[1]).reshape(1,-1))
expanded=np.vstack(rows)
```

```python
expanded
```

```python
galaxies=np.where(expanded!=0)
```

```python
galaxies
```

```python
galaxy_distances=[(abs(x2-x1)+abs(y2-y1)) for ((x1,y1),(x2,y2))  in combinations(zip(galaxies[0],galaxies[1]),2)]
sum(galaxy_distances)
```

```python
row_sum=np.sum(np_test,axis=1)
col_sum=np.sum(np_test,axis=0)

zero_rows=np.where(row_sum==0)[0]
zero_cols=np.where(col_sum==0)[0]

zero_rows,zero_cols
```

```python
zero_cols
```

```python
def compute_distance(r1,c1,r2,c2,emptyness=100):
    (rmin,rmax) = (r1,r2) if r1<r2 else (r2,r1)
    (cmin,cmax) = (c1,c2) if c1<c2 else (c2,c1)
    distance1=np.sum(zero_rows<rmax)-np.sum(zero_rows<rmin) 
    distance2=np.sum(zero_cols<cmax)-np.sum(zero_cols<cmin)
    distance3=rmax-rmin+cmax-cmin
    #return distance1,distance2,distance3, 
    return (emptyness-1)*(distance1+distance2)+distance3
    
```

```python
compute_distance(0,3,1,7,10)
```

```python
galaxies=np.where(np_test!=0)
galaxy_distances=[ compute_distance(x1,y1,x2,y2,100) for ((x1,y1),(x2,y2))  in combinations(zip(galaxies[0],galaxies[1]),2)]
sum(galaxy_distances)
```
